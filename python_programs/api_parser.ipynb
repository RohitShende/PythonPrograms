```python
import requests, re
import pandas as pd
from pyquery import PyQuery as pq
from lxml import etree
```

```python
url = "http://build-squid.eng.vmware.com/build/mts/release/bora-15698431/publish/nsx-unified-appliance/api_includes/api_all_methods.html"
res = requests.get(url)
data = res.content

d = pq(data)
anchors = [i.text for i in d('a')]
print(len(anchors))
rows = [i.text for i in d('th')]
data = [i.text_content().strip() for i in d('td')]
df = pd.DataFrame()
df[rows[0]] = data[0::2]
df[rows[1]] = data[1::2]
df['API Path'] = df['API Path'].apply(lambda x: [i.strip() for i in x.split('\n')])
df["Number of APIs"] = df["API Path"].apply(lambda x: len(x))
df
```
