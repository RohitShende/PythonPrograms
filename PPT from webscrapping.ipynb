```python
import requests
from pyquery import PyQuery
from IPython.display import Image, display
from pptx import Presentation
from pptx.util import Inches
from imageio import imread
import wget
import os
```

```python
url = 'https://www.femina.in/movies/a-fun-bollywood-game-to-play-with-family-if-youre-getting-bored-154909-1.html'
```

```python
resp = requests.get(url)
```

```python
resp.status_code
```

```python
html = resp.content.decode('utf-8')
```

```python
doc = PyQuery(html)
```

```python
photo_div = doc('div#photo_154909_3')
```

```python
image_url = photo_div('img')[0].attrib['data-original']
image_url
```

```python
prs=Presentation()
left=Inches(1)
top=Inches(1)
width  = Inches(8)
height = Inches(6)
```

```python
lyt=prs.slide_layouts[0] # choosing a slide layout
slide=prs.slides.add_slide(lyt) # adding a slide
title=slide.shapes.title # assigning a title
subtitle=slide.placeholders[1] # placeholder for subtitle
title.text="Lets Start With The Pictography Game !" # title
subtitle.text="We created this PPT using webscrapping and Python.. Exciting isnt it !" # subtitle
```

```python
for i in range(2, 22):
    photo_div = doc('div#photo_154909_{}'.format(i))
    image_url = photo_div('img')[0].attrib['data-original']
    print(i)
    
    
    display(Image(image_url))
    print(photo_div.text())
    print('\n\n')
    
    # download and add image to slide
    img = wget.download(image_url)
    blank_slide_layout = prs.slide_layouts[6]
    slide = prs.slides.add_slide(blank_slide_layout)
    pic = slide.shapes.add_picture(img, left, top, width, height)
    os.remove(img)
    
    lyt=prs.slide_layouts[0] # choosing a slide layout
    slide=prs.slides.add_slide(lyt) # adding a slide
    title=slide.shapes.title # assigning a title
    title.text=photo_div.text() # title
    
```

```python
prs.save("pictography_slides.pptx") # saving file
```
